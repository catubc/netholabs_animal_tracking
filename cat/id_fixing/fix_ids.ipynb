```python
import pandas as pd
import numpy as np

# --- Load and sort ---
df = pd.read_csv("/home/cat/data/yolo_csvs/cage_1_date_2025_08_29_hour_06_minute_48.csv")
df = df.sort_values("FRAME").reset_index(drop=True)

# --- Define constants ---
n_tracks = 10
n_frames = 1500
features = ["NOSE", "LEFT_EAR", "RIGHT_EAR", "LEFT_SIDE", "CENTER", "RIGHT_SIDE", "TAIL_BASE"]
n_features = len(features)

# --- Initialize array: [track, frame, feature, (x,y,p)] ---
tracks = np.full((n_tracks, n_frames, n_features, 3), np.nan)

# --- Fill array ---
for _, row in df.iterrows():
    frame = int(row["FRAME"])
    track = int(row["TRACK"])  # assuming TRACK starts from 0 or 1
    if track >= n_tracks or frame >= n_frames:
        continue  # skip extra tracks or frames
    
    for f_idx, f_name in enumerate(features):
        x = row[f"{f_name}_X"]
        y = row[f"{f_name}_Y"]
        p = row[f"{f_name}_P"]
        tracks[track - 1, frame, f_idx, :] = [x, y, p]  # adjust if TRACK starts at 1

```

```python
matched_tracks = np.full_like(tracks, np.nan)
```

```python
import numpy as np

n_tracks, n_frames, n_features, _ = tracks.shape

# Initialize matched array
matched_tracks = np.full_like(tracks, np.nan)

# Copy the first frame directly (no previous frame to compare)
matched_tracks[:, 0, :, :] = tracks[:, 0, :, :]

# Helper: compute centroid for each track (for matching)
def compute_centroid(frame_tracks):
    # Average of all visible feature points for that track
    centroids = np.nanmean(frame_tracks[:, :, :2], axis=1)  # shape (n_tracks, 2)
    return centroids

# --- Matching loop ---
for t in range(1, n_frames):
    prev_centroids = compute_centroid(matched_tracks[:, t - 1])
    curr_centroids = compute_centroid(tracks[:, t])

    # Identify valid (non-NaN) centroids
    prev_valid = ~np.isnan(prev_centroids[:, 0])
    curr_valid = ~np.isnan(curr_centroids[:, 0])

    prev_pts = prev_centroids[prev_valid]
    curr_pts = curr_centroids[curr_valid]

    if len(prev_pts) == 0 or len(curr_pts) == 0:
        continue

    # Compute distance matrix (n_prev_valid x n_curr_valid)
    dist = np.linalg.norm(prev_pts[:, None, :] - curr_pts[None, :, :], axis=2)

    # Greedy nearest-neighbor assignment
    for i, p_idx in enumerate(np.where(prev_valid)[0]):
        nearest_idx = np.argmin(dist[i])
        c_idx = np.where(curr_valid)[0][nearest_idx]

        matched_tracks[p_idx, t, :, :] = tracks[c_idx, t, :, :]

```

```python
import numpy as np
import matplotlib.pyplot as plt


def compute_centroid(frame_tracks: np.ndarray) -> np.ndarray:
    """
    Compute the centroid (mean x, y) across features for each frame of one track.
    Missing frames are filled with the previous valid centroid (forward fill).

    Parameters
    ----------
    frame_tracks : np.ndarray
        Array of shape (n_frames, n_features, 3), where the last dimension is (x, y, p).

    Returns
    -------
    np.ndarray
        Centroid positions of shape (n_frames, 2), containing (x, y) for each frame.
    """
    # Compute mean x,y per frame (ignoring NaNs)
    centroid = np.nanmean(frame_tracks[:, :, :2], axis=1)  # (n_frames, 2)

    # Forward-fill NaNs (replace missing frames with previous centroid)
    for i in range(1, centroid.shape[0]):
        if np.isnan(centroid[i, 0]) or np.isnan(centroid[i, 1]):
            centroid[i] = centroid[i - 1]

    return centroid



def plot_trajectory(animal_id: int,
                    ax: plt.Axes,
                    centroid: np.ndarray,
                    title: str = None,
                    cmap: str = "viridis"):
    """
    Plot the XY trajectory of a given animal, colored by time (frame index).

    Parameters
    ----------
    animal_id : int
        Index of the track (e.g., 0, 1, 2, ...).
    ax : matplotlib.axes.Axes
        Matplotlib axis object to draw the trajectory on.
    centroid : np.ndarray
        Centroid array of shape (n_frames, 2) for this animal.
    title : str, optional
        Title for the subplot.
    cmap : str, optional
        Matplotlib colormap name. Default is 'viridis'.
    """
    # Frame indices (for color)
    frames = np.arange(len(centroid))
    valid = ~np.isnan(centroid[:, 0])

    # Scatter with color by frame/time
    sc = ax.scatter(centroid[valid, 0],
                    centroid[valid, 1],
                    c=frames[valid],
                    cmap=cmap,
                    s=20,
                    alpha=0.9,
                    label=f'Animal {animal_id}')

    # Optional line connecting the points
    ax.plot(centroid[valid, 0], centroid[valid, 1],
            alpha=0.4, lw=1, color='gray')

    # Style the plot
    ax.set_xlabel('X position (px)')
    ax.set_ylabel('Y position (px)')
    ax.invert_yaxis()  # image coordinates
    if title:
        ax.set_title(title)
    ax.legend()

    # Add a colorbar for time
    cbar = plt.colorbar(sc, ax=ax)
    cbar.set_label('Frame index (time)')

    ax.set_ylim(0,230)
    ax.set_xlim(0,680)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

animal_id = 0
centroid_orig = compute_centroid(tracks[animal_id])
centroid_fixed = compute_centroid(matched_tracks[animal_id])

plot_trajectory(animal_id, axs[0], centroid_orig, "Original Trajectory")
plot_trajectory(animal_id, axs[1], centroid_fixed, "Matched Trajectory")


plt.tight_layout()
plt.show()



```
